```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Heart (1) .csv')
print(df.head())
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Age', y='MaxHR', hue='AHD', data=df)
plt.title('Scatter Plot: Age vs. MaxHR')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['ChestPain'].value_counts().plot(kind='bar', color='pink')
plt.title('Bar Plot: ChestPain Frequency')
plt.xlabel('ChestPain Type')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='AHD', y='Age', data=df)
plt.title('Box Plot: Age Distribution by AHD')
plt.xlabel('AHD')
plt.ylabel('Age')
plt.show()
```

```python
# Pie Chart: Proportion of AHD
plt.figure(figsize=(8, 6))
df['AHD'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'pink'])
plt.title('Pie Chart: Proportion of AHD')
plt.ylabel('')
plt.show()
```

```python
arr_Age = np.array(df["Age"])
sorted_Age = np.sort(arr_Age)
plt.figure(figsize=(8, 4))
plt.plot(sorted_Age, df['RestBP'], marker='o', linestyle='-', color='b')
plt.title('Line Chart: Age vs. RestBP')
plt.xlabel('Age')
plt.ylabel('RestBP')
plt.grid(True)
plt.show()
```
